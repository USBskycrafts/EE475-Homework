## 3.1

(a) 


$$g\prime(w)=log(w)-log(1-w)$$,

As the domain of $g\prime(w)$ is $(0,1)$, and it's monotonous, so there is only one stationary points at $w=0.5$

```python
def g1(w):
    return w*np.log10(w)+(1-w)*np.log10(1-w)
x = np.linspace(0, 1, 100)
y = list(map(g1, x))
plt.figure()
plt.plot(x, y)
plt.scatter(x=0.5, y=g1(0.5), c='r', label='minima')
```

And it's a minimum point.

(b)

$g\prime(w)=\frac{e^w}{(1+e^w)ln10} > 0$. When $w_0\rightarrow -\infty$, $g(w_0) \rightarrow0$

```python
def g2(w):
    return np.log10(1+np.exp(w))
x = np.linspace(-20, 20, 1000)
y = list(map(g2, x))
plt.plot(x, y)
```

so there's only a saddle point at $x=-\infty$

(c)

$$g\prime(w)={tanh}(w)+w(1-{tanh}^2(w))$$

it's easy to notice that there is a stationary point at $w=0$, now we prove that there is no other stationary point.

for $w > 0$, $tanh(w) \in (0, 1)$, $1-tanh^2(w) \in (0, 1)$, so $g\prime(w) > 0$, there is no stationary point.

for $w < 0$, assume that there is a stationary point at $w = w_0$

$\rightarrow tanh(w_0)+ w_0(1-tanh^2(w_0)) = 0$

$\rightarrow \frac{1}{w_0}=tanh(w_0)-\frac{1}{tanh(w_0)}$

let $h(w) = tanh(w)-\frac{1}{tanh(w)}$, it's easy to prove that $h(w)$ is a monotonically increasing function. when $w \rightarrow -\infty$, $h(w) \rightarrow -1 + 1 = 0$. However, $\frac{1}{w_0} \in (-\infty, 0)$, so $w_0$ does not exist.

```python
def g3(w):
    return w*np.tanh(w)
x = np.linspace(-20, 20, 1000)
y = list(map(g3, x))
plt.plot(x, y)
plt.scatter(x=0, y=g3(0), c='r', label='minima')
```

the stationary point at $w=0$ is a minina.

(d)

$$g\prime(\boldsymbol{w})=\frac{1}{2}\cdot2\boldsymbol{C}\boldsymbol{w} + \boldsymbol{b} =\boldsymbol{C}\boldsymbol{w} + \boldsymbol{b}$$

When $g\prime(\boldsymbol{w})=0$, we get $\boldsymbol{w}=[-\frac{1}{5}, -\frac{2}{5}]^T$

```python
def g4(w : np.array):
    w = np.expand_dims(w, axis=0)
    w = w.T
    return 0.5*w.T@np.array([[2,1],[1,3]])@w+np.array([[1, 1]])@w
x = np.linspace(-1, 1)
y = np.linspace(-1, 1)
grid = np.meshgrid(x, y)
arr = np.transpose(grid, (1,2,0))
g = np.empty((arr.shape[0], arr.shape[1]))
for i in range(g.shape[0]):
    for j in range(g.shape[1]):
        g[i][j] = g4(arr[i][j]).squeeze()
X, Y = grid
plt.pcolor(X, Y, g)
plt.xlabel('w1')
plt.ylabel('w2')
plt.colorbar()
```

we can see that there is a minima at $(-0.25, -0.4)$

$$\frac{\partial g(\boldsymbol{w})}{\partial \boldsymbol{w}}=\frac{2(C\boldsymbol{w}) \boldsymbol{w}^T\boldsymbol{w}-\boldsymbol{w}^T\boldsymbol{C}\boldsymbol{w}\cdot 2\boldsymbol{w}_j)}{(\boldsymbol{w}^T\boldsymbol{w})^2}=\frac{2}{\boldsymbol{w}^T\boldsymbol{w}}(\boldsymbol{Cw}-\frac{\boldsymbol{w}^T\boldsymbol{C}\boldsymbol{w}}{(\boldsymbol{w}^T\boldsymbol{w})^2}\boldsymbol{w})$$
notice that
$$g(\boldsymbol{w})=\frac{\boldsymbol{w}^T\boldsymbol{C}\boldsymbol{w}}{(\boldsymbol{w}^T\boldsymbol{w})^2}$$
we can rewrite it as
$$\frac{\partial g(\boldsymbol{w})}{\partial \boldsymbol{w}}=\frac{2}{\boldsymbol{w}^T\boldsymbol{w}}(\boldsymbol{Cw}-g(\boldsymbol{w})\boldsymbol{w})$$

if $g\prime(\boldsymbol{w_0})=0$, then $\boldsymbol{Cw_0}=g(\boldsymbol{w_0})\boldsymbol{w_0}$, which means that the stationary points of $g(\boldsymbol{w})$ are all the eigenvectors of $\boldsymbol{C}$
