```python
# import basic libraries and autograd wrapped numpy
import numpy as np
import copy
import matplotlib.pyplot as plt


# this is needed to compensate for matplotlib notebook's tendancy to blow up images when plotted inline
%matplotlib inline
from matplotlib import rcParams
rcParams['figure.autolayout'] = True
```

```python
def quadratic(x):
    return np.dot(x.T,x)
```

N is dimension and P is the sampling number

```python
N = 100
P = 100
y = [0] * N

for i in range(1, N + 1):
    x_arr = [np.linspace(-1, 1, P) for _ in range(P)]
    x = np.meshgrid(x_arr)
    x = np.vstack(list(map(np.ravel,x))).T
    x_min = min(x, key=lambda x : quadratic(x))
    y[i - 1] = quadratic(x_min)
    #print(f"demension is {i}, y-min is {y[i - 1]}")
```

```python
x = [i for i in range(1, N + 1)]
plt.figure()
plt.plot(x, y)
plt.xlabel("N")
plt.ylabel("mins{g(w)}")
plt.show()
```

```python
def g(x, y):
    return 100 * (y - x**2)**2 + (x - 1)**2
```

```python
def fixed_step_opt():
    K = 50
    P = 1000
    w = [-2, -2]
    a = 1
    x = [-2]
    y = [-2]
    color = [0]
    for k in range(K):
        flag = False
        u, v = 0, 0
        flag = False
        w_tmp = [w[0], w[1]]
        for _ in range(P):
            t = np.random.rand() * 2 * np.pi
            u = np.cos(t)
            v = np.sin(t)
            if g(w[0] + u, w[1] + v) < g(w_tmp[0], w_tmp[1]):
                w_tmp[0] = w[0] + u
                w_tmp[1] = w[1] + v
                flag = True
        if flag == False:
            break
        #print((w_tmp[0] - w[0]) ** 2 + (w_tmp[1] - w[1])**2)
        w[0] = w_tmp[0]
        w[1] = w_tmp[1]
        #print(w)
        x.append(w[0])
        y.append(w[1])
        color.append(k)
    print(g(x[-1], y[-1]))
    return (x, y, color)
```

```python
x, y, color = fixed_step_opt()
plt.figure()
plt.scatter(x, y, c=color, cmap='viridis')
```

```python
def dimin_step_opt():
    K = 50
    P = 1000
    w = [-2, -2]
    a = 1
    x = [-2]
    y = [-2]
    color = [0]
    for k in range(1, K + 1):
        flag = False
        u, v = 0, 0
        flag = False
        w_tmp = [w[0], w[1]]
        for _ in range(P):
            t = np.random.rand() * 2 * np.pi
            u = np.cos(t)
            v = np.sin(t)
            if g(w[0] + 1/k * u, w[1] + 1/k * v) < g(w_tmp[0], w_tmp[1]):
                w_tmp[0] = w[0] + 1/k * u
                w_tmp[1] = w[1] + 1/k * v
                flag = True
                break
        if flag == False:
            #print(g(w[0], w[1]))
            break
        w[0] = w_tmp[0]
        w[1] = w_tmp[1]
        #print(w)
        x.append(w[0])
        y.append(w[1])
        color.append(k)
    print(g(x[-1], y[-1]))
    return (x, y, color)
```

```python
x, y, color = dimin_step_opt()
plt.figure()
plt.scatter(x, y, c=color, cmap='viridis')
```

**conlusion**
1. in this suituation, fixed steplength search get the smaller minima
2. K=50 is far more enough for fixed steplength search, but is not enough for diminish steplength search

### 2.5 (a)

the descent probability can be calculated as
$$P = \frac{the \ length \ of \ thick \ black \ line}{the \ perimeter \ of \ the \ unit \ circle \ centered \ at \ \mathbf{w_0}}$$
$P = \frac{\frac{2}{3}\pi}{2\pi}=\frac{1}{3}$ when N = 2

Notice that the length of thick black line is guaranteed to be smaller than half of the perimeter of the blue circle, so $P < \frac{ \frac{1}{2}\times 2\pi \frac{\sqrt{3}}{2} }{2\pi}=\frac{\sqrt{3}}{4}$

### 2.5 (b)

Now we just need to transform circle into hypersphere in high-dimensional space, so
$$P=\frac{\frac{1}{2}\times {\frac{\sqrt{3}}{2}}^{N-1}\pi}{\pi}=\frac{1}{2}({\frac{\sqrt{3}}{2}})^{N-1}$$

```python
def h(x, y):
    return x**2 + y**2 + 2
```

```python
def random_search():
    P = 1000
    w = [3, 4]
    x, y = [3], [4]
    color = [0]
    a = 1
    K = 5
    for k in range(K):
        flag = False
        u, v = 0, 0
        flag = False
        for _ in range(P):
            t = np.random.rand() * 2 * np.pi
            u = np.cos(t)
            v = np.sin(t)
            if h(w[0] + u, w[1] + v) < h(w[0], w[1]):
                flag = True
                break
        if flag == False:
            break
        w[0] = w[0] + u
        w[1] = w[1] + v
        x.append(w[0])
        y.append(w[1])
        color.append(k)
    print(h(x[-1], y[-1]))
    return (x, y, color)
```

```python
x, y, color = random_search()
plt.figure()
plt.scatter(x, y, c=color, cmap='viridis')
```

```python
def coordinate_search():
    P = 1000
    w = [3, 4]
    x, y = [3], [4]
    color = [0]
    a = 1
    K = 7
    for k in range(K):
        flag = False
        u, v = 0, 0
        flag = False
        w_tmp = [w[0], w[1]]
        for u, v in [[-1, 0], [0, 1], [1, 0], [0, -1]]:
            if h(w[0] + u, w[1] + v) < h(w_tmp[0], w_tmp[1]):
                w_tmp[0], w_tmp[1] = w[0] + u, w[1] + v
                flag = True
        if flag == False:
            break
        w[0] = w_tmp[0]
        w[1] = w_tmp[1]
        x.append(w[0])
        y.append(w[1])
        color.append(k)
    print(h(x[-1], y[-1]))
    return (x, y, color)
```

```python
x, y, color = coordinate_search()
plt.figure()
plt.scatter(x, y, c=color, cmap='viridis')
```

### the description of behavior:

1. coordinate search descending along the coordinate axes, while random search is not.
2. coordinate search reaches the minimal point, while random search is just close to the minima.
